#joint Distribution 


```python
x=np.random.normal(size=1000)
y=np.random.normal(size=1000)
```

```python
df=pd.DataFrame({'x':x,'y':y})
```

```python
sns.jointplot(df.x,df.y);
```

```python
sns.jointplot(df.x,df.y,kind='kde',shade=True);
```

```python
sns.jointplot(d.x,d.y);
```

```python
x=np.random.normal(size=1000)
y=5*x*np.random.normal(size=1000)/5
```

```python
df=pd.DataFrame({'x':x,'y':y})
```

```python
sns.jointplot(df.x,df.y);
```
